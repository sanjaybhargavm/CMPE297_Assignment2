```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained(
        "jinaai/starcoder-1b-textbook", device_map='auto'
    )

tokenizer = AutoTokenizer.from_pretrained("jinaai/starcoder-1b-textbook")
```

```python
prompt = '''
Given an array of strings strs, identify all the palindromes and group them together.
A palindrome is a word, phrase, number, or other sequences of characters that reads the same forward and backward
(ignoring spaces, punctuation, and capitalization). Return the palindromes in a list."

Example 1
Input: strs = ["level", "deed", "rotor", "noon", "apple", "peach"]
Output: ["level", "deed", "rotor", "noon"]

def is_palindrome
'''


inputs = tokenizer(prompt.rstrip(), return_tensors="pt").to("cuda")

generation_output = model.generate(
    **inputs,
    max_new_tokens=256,
    eos_token_id=tokenizer.eos_token_id,
    return_dict_in_generate=True,
)

s = generation_output.sequences[0]
output = tokenizer.decode(s, skip_special_tokens=True)

print(output)
```
